```python
import random

# ASCII Hangman Stages
hangman_pics = [
    """
       ------
       |    |
       |
       |
       |
       |
    --------
    """,
    """
       ------
       |    |
       |    O
       |
       |
       |
    --------
    """,
    """
       ------
       |    |
       |    O
       |    |
       |
       |
    --------
    """,
    """
       ------
       |    |
       |    O
       |   /|
       |
       |
    --------
    """,
    """
       ------
       |    |
       |    O
       |   /|\\
       |
       |
    --------
    """,
    """
       ------
       |    |
       |    O
       |   /|\\
       |   /
       |
    --------
    """,
    """
       ------
       |    |
       |    O
       |   /|\\
       |   / \\
       |
    --------
    """
]

def get_word():
    words = ["python", "hangman", "challenge", "developer", "programming"]
    return random.choice(words)

def display(word, guessed_letters):
    return " ".join([letter if letter in guessed_letters else "_" for letter in word])

def hangman():
    word = get_word()
    guessed_letters = set()
    attempts = 6

    print("ğŸ‰ Welcome to Hangman! ğŸ‰")

    while attempts > 0:
        print(hangman_pics[6 - attempts])  # Show hangman stage
        print(display(word, guessed_letters))

        guess = input("ğŸ”¤ Guess a letter: ").lower()

        if len(guess) != 1 or not guess.isalpha():
            print("â— Please enter a single valid letter.")
            continue

        if guess in guessed_letters:
            print("âš ï¸ You already guessed that letter!")
        elif guess in word:
            guessed_letters.add(guess)
            print("âœ… Good guess!")
            if set(word) == guessed_letters:
                print(f"ğŸ‰ Congratulations! You guessed the word: {word} ğŸ‰")
                return
        else:
            attempts -= 1
            print(f"âŒ Wrong guess! Attempts left: {attempts}")

    print(hangman_pics[6])  # Show final hangman state
    print(f"ğŸ’€ Game Over! The word was: {word}")

hangman()
```
